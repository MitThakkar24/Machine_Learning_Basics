# Standardization

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
total_bill = list(df['total_bill'])
```

```python
import numpy as np
mean = np.mean(total_bill)
std = np.std(total_bill)
```

```python
mean,std
```

```python
normalized_data=[]
for i in total_bill:
    z_score = (i-mean)/std
    normalized_data.append(z_score) # Z Score on Every Element
    
    
normalized_data
```

```python
sns.histplot(total_bill)
```

```python
sns.histplot(normalized_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df.head()
```

```python
scaler.fit(df[['total_bill','tip']])
```

```python
transformed_data = scaler.transform(df[['total_bill','tip']])
```

```python
import pandas as pd

pd.DataFrame(transformed_data, columns=['total_bill','tip'])
```
